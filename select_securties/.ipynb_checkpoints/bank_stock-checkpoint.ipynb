```python
def get_all_securties():
    df_all = ts.get_industry_classified()
    df_finance = df_all[df_all['c_name'] == '金融行业']
    return df_finance[df_finance['name'].str.contains('证')]

get_all_securties()
```

```python
ts.get_hist_data('600848').sample(10) #一次性获取全部日k线数据
```

- date：日期
- open：开盘价

- price_change：价格变动
- p_change：涨跌幅
- ma5：5日均价
- ma10：10日均价
- ma20:20日均价
- v_ma5:5日均量
- v_ma10:10日均量
